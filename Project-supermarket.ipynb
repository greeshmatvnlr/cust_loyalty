# Gender

```python
ax = sns.countplot(x = 'Male/Female',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

# Loyalty card users Vs non-users

```python
ax = sns.countplot(x = 'Did the participants used loyalty card',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

# Loyalty voucher user Vs non-users 

```python
bx = sns.countplot(x = 'Did the customer used loyalty voucher for shopping?',data = df)
for bars in bx.containers:
    bx.bar_label(bars)
```

# Customers purchased to use the vouchers Vs non-users

```python
cx = sns.countplot(x = 'Did the customer purchased any item to use the loyalty voucher?',data = df)
for bars in cx.containers:
    cx.bar_label(bars)
```

# How often people visit store A

```python
dx = sns.countplot(x = 'How often visit at supermarket A',data = df)
for bars in dx.containers:
    dx.bar_label(bars)
```

# Other stores visisted by the customers of store A

```python
ex = sns.countplot(x = 'Other supermarket often visits by the customer',data = df)
for bars in ex.containers:
    ex.bar_label(bars)
```

# Total Bill amount Vs gender

```python
bill_gen = df.groupby(['Male/Female'], as_index=False)['Bill(Â£)'].sum().sort_values(by='Bill(Â£)', ascending=False)

fx=sns.barplot(x = 'Male/Female',y= 'Bill(Â£)' ,data = bill_gen)

for bars in fx.containers:
    fx.bar_label(bars)
```

# Total Bill paid by loyalty card users Vs non users

```python
bill_loyalty = df.groupby(['Did the participants used loyalty card'], as_index=False)['Bill(Â£)'].sum().sort_values(by='Bill(Â£)', ascending=False)

gx=sns.barplot(x = 'Did the participants used loyalty card',y= 'Bill(Â£)' ,data = bill_loyalty)

for bars in gx.containers:
    gx.bar_label(bars)
```

# Total bill amount paid by voucher users vs non users

```python
bill_voucher = df.groupby(['Did the customer used loyalty voucher for shopping?'], as_index=False)['Bill(Â£)'].sum().sort_values(by='Bill(Â£)', ascending=False)

hx=sns.barplot(x = 'Did the customer used loyalty voucher for shopping?',y= 'Bill(Â£)' ,data = bill_voucher)

for bars in hx.containers:
    hx.bar_label(bars)
```

```python
bill_voucher_for_item = df.groupby(['Did the customer purchased any item to use the loyalty voucher?'], as_index=False)['Bill(Â£)'].sum().sort_values(by='Bill(Â£)', ascending=False)

ix=sns.barplot(x = 'Did the customer purchased any item to use the loyalty voucher?',y= 'Bill(Â£)' ,data = bill_voucher_for_item)

for bars in ix.containers:
    ix.bar_label(bars)
```
